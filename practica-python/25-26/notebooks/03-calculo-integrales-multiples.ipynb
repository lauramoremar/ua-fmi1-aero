# Cálculo de integrales múltiples



## Ejemplo 1

$$ \int_{0}^{1} \int_{4x}^{4} (x+y) \, dy \, dx = 6 $$

```python
# Importamos el paquete sympy para realizar cálculo simbólico
import sympy as sym

#Definimos las variables "x" e "y" como símbolos
x, y = sym.symbols("x y")

#Definimos la función a integrar
f = x + y

```

```python
#Definimos la integral
integral = sym.Integral(f, (y, 4*x, 4), (x, 0, 1))

display(integral)  #sym.pprint(integral)
print("Resultado:")
display(integral.doit())
```

## Ejemplo 2

Calcular la integral

$$ \int_{0}^{\pi/ 4} \int_{0}^{2} \rho^2 \cos \omega \, d\rho \, d\omega = \dfrac{4 \sqrt{2}}{3} $$

```python
#Definimos las variables r y w:
r, w =sym.symbols("r w")

#Definimos la función a integrar
f = r**2*sym.cos(w)
f
```

```python
#Definimos la integral
integral = sym.Integral(f, (r, 0, 2), (w, 0, sym.pi/4))

display(integral)  #sym.pprint(integral)
print("Resultado:")
display(integral.doit())
```

## Ejemplo 3

Calcular la integral

 $$ \int_{0}^{4} \int_{0}^{\pi/2} \int_{0}^{1-x} x \cos y \, dz \, dy \, dx = - \dfrac{40}{3} $$

```python
#Definimos las variables "x" e "y" como símbolos
z =sym.symbols("z")

#Definimos la función a integrar
f = x*sym.cos(y)

#Definimos la integral
integral = sym.Integral(f, (z, 0, 1-x), (y, 0, sym.pi/2), (x, 0, 4))

display(integral)  #sym.pprint(integral)
print("Resultado:")
display(integral.doit())
```

## Ejemplo 4
Calcular la integral 
$$
\int_{0}^{2\pi} \int_{0}^{R} \int_{0}^{4\rho} z \rho \, dz \, d\rho \, d\alpha = 4 \pi R^4
$$

```python
#Definimos las variables r, w, z y la constante R:
r, a, z, R =sym.symbols("r a z R")

#Definimos la función a integrar
f = z*r

#Definimos la integral
integral = sym.Integral(f, (z, 0, 4*r), (r, 0, R), (a, 0, 2*sym.pi))

display(integral)  #sym.pprint(integral)
print("Resultado:")
display(integral.doit())
```
